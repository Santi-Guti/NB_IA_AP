## 2. Elimine registros que tengan NA en la columna "pm2.5"

```python
datos_limpios = datos.dropna(subset=["pm2.5"], axis=0, inplace=False)
datos_limpios = datos_limpios[["year", "month", "day", "hour", "pm2.5", "DEWP", "TEMP", "PRES", "cbwd", "Iws", "Is", "Ir"]]
display(datos_limpios)
```

## 3. Realice un describe de todas las variables por año. Además de realizar un gráfico como boxplot con el que pueda comparar los resultados dados.

```python
fig, axs = plt.subplots(2, 2)

axs[0, 0].boxplot(datos[["pm2.5"]])
axs[0, 0].set_title('pm2.5')

axs[0, 1].boxplot(datos[["DEWP"]])
axs[0, 1].set_title('DEWP')

axs[1, 0].boxplot(datos[["TEMP"]])
axs[1, 0].set_title('TEMP')

axs[1, 1].boxplot(datos[["PRES"]])
axs[1, 1].set_title('PRES')


fig.subplots_adjust(left=0.08, right=0.98, bottom=0.05, top=0.9,
                    hspace=0.4, wspace=0.3)

plt.show()
```

## 4. Para finalizar, agrupe por año el valor de la variable "pm2.5", calcule el promedio y ordene la base de datos de manera ascendente por el promedio.

```python
datos_agr_asc = (datos.groupby("year")[["pm2.5"]]
                 .mean()
                 .sort_values("pm2.5", ascending=True))
datos_agr_asc
```
